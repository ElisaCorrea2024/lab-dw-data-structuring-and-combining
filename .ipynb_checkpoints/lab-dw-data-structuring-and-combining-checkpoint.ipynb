## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd
from IPython.display import display  # Para mostrar el DataFrame como tabla

# URLs de los archivos CSV
url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
url2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'

# Leer los archivos CSV en DataFrames
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)

# Combinar los datasets
combined_df = pd.concat([df1, df2, df3], ignore_index=True)

# Definir una función de limpieza
def clean_data(df):
    # Limpiar los nombres de las columnas
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    
    # Calcular el porcentaje de valores nulos por columna
    null_percentage = df.isnull().mean() * 100
    print("Porcentaje de valores nulos por columna:")
    print(null_percentage)
    
    # Eliminar columnas con más del 50% de valores nulos
    threshold = 0.5 * df.shape[0]  # 50% del número total de filas
    df_cleaned = df.dropna(thresh=threshold, axis=1)  # Mantener columnas con al menos el 50% de datos
    
    return df_cleaned

# Aplicar la función de limpieza al DataFrame combinado
cleaned_df = clean_data(combined_df)

# Mostrar el DataFrame limpio como tabla y las primeras 10 filas
display(cleaned_df.head(10))
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
import pandas as pd
from IPython.display import display  # Para mostrar el DataFrame como tabla

# Cargar el dataset
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
df = pd.read_csv(url)

# Mostrar las primeras filas del DataFrame
print("Primeras filas del DataFrame:")
display(df.head())
```

```python
# Información general del DataFrame
print("\nInformación del DataFrame:")
print(df.info())

# Resumen estadístico de las columnas numéricas
print("\nResumen estadístico:")
print(df.describe())

# Verificar la cantidad de valores nulos
print("\nCantidad de valores nulos por columna:")
print(df.isnull().sum())
```

```python
# Limpieza de columnas (puedes ajustar según sea necesario)
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Eliminación de columnas con más del 50% de nulos
threshold = 0.5 * df.shape[0]
df_cleaned = df.dropna(thresh=threshold, axis=1)

# Verificar la cantidad de valores nulos después de la limpieza
print("\nCantidad de valores nulos después de la limpieza:")
print(df_cleaned.isnull().sum())
```

```python
# Ejemplo de convertir variables categóricas en variables dummy
df_dummies = pd.get_dummies(df_cleaned, drop_first=True)

# Mostrar las primeras filas del DataFrame estructurado
print("\nPrimeras filas del DataFrame estructurado:")
print(df_dummies.head())
```
